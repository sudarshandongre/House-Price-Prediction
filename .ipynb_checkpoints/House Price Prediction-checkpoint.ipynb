# Try 2

```python
data2  = data.copy()
```

```python
data2.drop(columns=['Price_sqft', 'type_of_building', 'neworold', 'Bathrooms', 'longitude', 'latitude', 'Status'], inplace=True)
```

```python
data2
```

```python
X2 = data2.drop(columns=['price'])
Y2 = data2['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2,Y2, test_size=0.2, random_state=0)
```

```python
column_trans2 = make_column_transformer((OneHotEncoder(), ['Address']), remainder='passthrough')
```

```python
scalar2  = StandardScaler(with_mean=False)
```

```python
lr2 = LinearRegression()
```

```python
pipe2 = make_pipeline(column_trans2,scalar2,lr2)
```

```python
pipe2.fit(X2_train,Y2_train)
```

```python
Y2_pred_lr = pipe2.predict(X2_test)
```

```python
r2_score(Y2_test,Y2_pred_lr)
```

```python
def convert_to_lakhs(df):
  """
  This function converts values in a pandas dataframe column from scientific notation 
  to lakhs (divide by 100,000).

  Args:
      df (pandas.DataFrame): The dataframe containing the column to convert.

  Returns:
      pandas.DataFrame: The dataframe with the converted column.
  """
  # Select the column 'price'
  price_col = df['price']

  # Check if the column is not of string type (assuming it's numeric)
  if not pd.api.types.is_string_dtype(price_col):
    # Convert the column to string type to enable splitting on 'e'
    price_col = price_col.astype(str)

  # Split the mantissa and exponent, handle cases where 'e+' is missing
  try:
    mantissa, exponent = price_col.str.split('e+', expand=True).values.T
  except ValueError:
    # If splitting fails (no 'e+'), assume no exponent and set to 0
    mantissa = price_col.values
    exponent = pd.Series([0] * len(price_col))

  # Convert mantissa to float and exponent to int
  mantissa = mantissa.astype(float)
  exponent = exponent.astype(int)

  # Invert the exponent (since we're dividing by 10^5)
  new_exponent = -5

  # Combine mantissa and new exponent with 'e+'
  converted_values = mantissa * (10**new_exponent)

  # Update the 'price' column with converted values
  df['price'] = converted_values

  return df
```

```python
data3 = convert_to_lakhs(data3.copy())
```
